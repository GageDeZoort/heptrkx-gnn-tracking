# Message passing neural network segment classifier evaluation

In this notebook we evaluate the results of training a graph network on the TrackML dataset (filtered/cleaned) to classify hit-pair track segments.

The model was trained using the trainSegmentClassifier.py script.

```python
# Compatibility
from __future__ import print_function
from __future__ import division
from __future__ import absolute_import
```

```python
# System imports
import os
```

```python
# Externals
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Magics
%matplotlib notebook
```

## Load the model and data

```python
train_dir = '/bigdata/shared/HEP.TrkX/GNNSegClf_trkml1GeV_002'
```

```python
ls $train_dir
```

```python
model_file = os.path.join(train_dir, 'model')
losses_file = os.path.join(train_dir, 'losses.npz')
```

```python
losses_data = np.load(losses_file)
train_losses = losses_data['train_losses']
valid_losses = losses_data['valid_losses']
```

```python
# Plot the loss
plt.figure()
plt.plot(train_losses, label='training set')
plt.plot(valid_losses, label='validation set')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend(loc=0);
```
